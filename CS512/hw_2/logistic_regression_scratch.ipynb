## Question 3.1

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CS512/hw_2/titanicdata.csv')
print(data.head())
```

```python
#feature normalization
mms = MinMaxScaler()
data[['Age']] = mms.fit_transform(data[['Age']])
data.head()
```

```python
data.values
```

```python
Y = data.values[:,0]
X = data.values[:,1:4]
```

```python
def sigmoid_func(scores):
    return 1 / (1 + np.exp(-scores))

def log_likelihood(my_features, target_variable, weights):
    scores = np.dot(my_features, weights)
    ll = np.sum( target_variable*scores - np.log(1 + np.exp(scores)) )
    return ll

def logistic_regression(my_features, target_variable, num_of_steps, learning_rate, add_intercept = False):
    if add_intercept: # if in case intercept is added
        intercept = np.ones((my_features.shape[0], 1))
        my_features = np.hstack((intercept, my_features)) # this is there to stack the sequence of my input arrays

    weights = np.zeros(my_features.shape[1])

    for i in range(num_of_steps): # for every step
        scores = np.dot(my_features, weights)
        predictions = sigmoid_func(scores)

        # This bunc of code will update weights with gradient ('grad' in my case) for every step
        output_error = target_variable - predictions
        grad = np.dot(my_features.T, output_error)
        weights += learning_rate * grad
        
        # After calculations, this will print the log-likelihood based on my condition
        if i % 10000 == 0:
            print(log_likelihood(my_features, target_variable, weights))
        
    return weights
```

```python
weights = logistic_regression(X, Y, num_of_steps = 30000, learning_rate = 4e-5, add_intercept=True)
```

```python
data_with_intercept = np.hstack((np.ones((X.shape[0], 1)),X))
final_scores = np.dot(data_with_intercept, weights)
preds = np.round(sigmoid_func(final_scores))

print('Accuracy from scratch: {0}'.format((preds == Y).sum().astype(float) / len(preds)))
#SKlearn
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# Fit the model to the training data
model.fit(X, Y)
# Make predictions on the test data
y_pred = model.predict(X)
# Evaluate the model's accuracy
accuracy = model.score(X, Y)
print(f'Accuracy with SKlearn: {accuracy:.3}')
```
